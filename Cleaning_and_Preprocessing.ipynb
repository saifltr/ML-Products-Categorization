```python
import pandas as pd
import os
```

<h1> Concatinating all the csv files </h1>

```python
# Directory where all the different csv files present
directory = 'data/'

csv_files = [
    'asos.csv',
    'adidas.csv',
    'nike.csv',
    'sportsdirect.csv',
    'argos.csv',
    'decathlon.csv',
    'ao.csv',
    'currys.csv',
    'sephora.csv',
    'next.csv',
    'amazon.csv'
]

# Iterates through the list of CSV filenames, attempts to read each CSV file
for file in csv_files:
    file_path = os.path.join(directory, file)
    print(f"Columns in {file}:")
    try:
        df = pd.read_csv(file_path)
        print(df.columns.tolist())
    except FileNotFoundError:
        print(f"File '{file}' not found.")
    print()
```

```python
dfs = []

#iterates through a list of CSV filenames, reads each CSV file using pandas, selects specific columns from the DataFrame, and appends the modified DataFrame to a list called dfs
for file in csv_files:
    file_path = os.path.join(directory, file)
    try:
        # Read the CSV file
        df = pd.read_csv(file_path, low_memory=False)
        # Select only the columns that exist in the DataFrame and are required
        columns_to_keep = [col for col in df.columns if col in ['product_title', 'product_category', 
                                                                 'original_price', 'selling_price', 
                                                                 'discount_percentage', 'product_rating', 
                                                                 'meta_data_1', 'meta_data_2', 'meta_data_3', 
                                                                 'review_count', 'spider_name']]
        df = pd.read_csv(file_path, usecols=columns_to_keep, low_memory=False)
        dfs.append(df)
    except FileNotFoundError:
        print(f"File '{file}' not found.")

result_df = pd.concat(dfs, ignore_index=True)
result_df.to_csv('products.csv', index=False)

print("Concatenation completed. The result is saved in 'products.csv'.")
```

```python
from bs4 import BeautifulSoup
import numpy as np

# Define a function to extract product description from meta_data_1
def extract_product_description(html):
    if pd.isnull(html):
        return np.nan
    soup = BeautifulSoup(html, 'html.parser')
    product_description = soup.find('h2', text='Product Description')
    if product_description:
        details_table = product_description.find_next('table')
        if details_table:
            details = {}
            for row in details_table.find_all('tr'):
                cells = row.find_all('td')
                if len(cells) == 2:
                    key = cells[0].text.strip()
                    value = cells[1].text.strip()
                    details[key] = value
            return details
    return np.nan

# Apply the function to the meta_data_1 column only for rows where spider_name is 'amazon'
df.loc[df['spider_name'] == 'amazon', 'meta_data_1'] = df.loc[df['spider_name'] == 'amazon', 'meta_data_1'].apply(extract_product_description)
```

```python
# Replace 'meta_data1' and 'spider_name' with your actual column names

# Iterate over rows in the DataFrame
for index, row in df.iterrows():
    # Check if the spider_name is 'amazon'
    if row['spider_name'] == 'amazon':
        # Remove single quotes and curly braces from the meta_data1 value
        df.at[index, 'meta_data_1'] = str(row['meta_data_1']).replace("'", "").replace("{", "").replace("}", "")
```
